```python
def buildSWTable(seq1, seq2, match = 1, mismatch = -1, gap = -1):
  import pandas as pd
  print('Seq1: ',seq1)
  print('Seq2: ',seq2,"\n")
  table = [[0]*(len(seq2)+1) for i in range(len(seq1)+1)] # create initial table with 0
  print("The initial alignment table is: ")
  pd_table = pd.DataFrame(table)
  print(pd_table)
  print("\n")

  ## start update the first column and the first row, for local alignemnt first row and column is 0

  for i in range(len(seq2)+1): # update first row
    table[0][i] = 0
  for i in range(len(seq1)+1): # update first column
    table[i][0] = 0

  ## build the alignment table

  for rowid in range (1,len(seq1)+1):
    for colid in range(1,len(seq2)+1):
      if seq1[rowid-1] == seq2[colid-1]:
        case1 = table[rowid-1][colid-1] + match
      else:
        if table[rowid-1][colid-1] > 0: # update the 0 into the SWTable, Bc value can't be smaller than 0
          case1 = table[rowid-1][colid-1] + mismatch
        else:
          case1 = 0 

      case2 = table[rowid-1][colid]+gap

      case3 = table[rowid][colid-1]+gap

      table[rowid][colid] = max([case1,case2,case3])
  print("The final alignment table is:")
  pd_table = pd.DataFrame(table)
  print(pd_table)
  print("\n")

 
```

```python
def SWTraceBAck(seq1, seq2, match = 1, mismatch = -1, gap = -1):
  import pandas as pd
  print('Seq1: ',seq1)
  print('Seq2: ',seq2,"\n")
  Max = 0
  temp_xy=[]
  table = [[0]*(len(seq2)+1) for i in range(len(seq1)+1)]
  ## start update the first column and the first row, for local alignemnt first row and column is 0

  for i in range(len(seq2)+1): # update first row
    table[0][i] = 0
  for i in range(len(seq1)+1): # update first column
    table[i][0] = 0

  ## build the alignment table

  for rowid in range (1,len(seq1)+1):
    for colid in range(1,len(seq2)+1):
      if seq1[rowid-1] == seq2[colid-1]:
        case1 = table[rowid-1][colid-1] + match
      else:
        if table[rowid-1][colid-1] > 0: # update the 0 into the SWTable, Bc value can't be smaller than 0
          case1 = table[rowid-1][colid-1] + mismatch
        else:
          case1 = 0 

      case2 = table[rowid-1][colid]+gap

      case3 = table[rowid][colid-1]+gap

      table[rowid][colid] = max([case1,case2,case3])
      temp_max = max([case1,case2,case3])

      ##  during iteration, find where the max value locate in the table, we need to update the bottom right one
      ## that's why need to add = sign, when meet the max but not at bottom right --> keep update the row and column
      if temp_max >=  Max: 
        temp_xy.clear()
        Max = temp_max
        temp_xy.append(rowid)
        temp_xy.append(colid)

  print("The final alignment table is:")
  pd_table = pd.DataFrame(table)
  print(pd_table)
  print("\n")
  print("The score of optimal alignment is:", Max)
  print("Locate at row,col: ", temp_xy)

  ## the best alignmnet
  
  col_start = temp_xy[1]
  row_start = temp_xy[0]
  list1=[]
  blank=[]
  list2=[]
  while Max > 0: #if max ==0 than stop alignment
    if seq1[row_start-1] == seq2[col_start-1]:
      case1 = table[row_start - 1][col_start - 1] + match
    else:
      case1 = table[row_start - 1][col_start - 1] + mismatch

    case2 = table[row_start - 1][col_start] + gap

    case3 = table[row_start][col_start - 1] + gap


    if case1 == table[row_start][col_start]:
      list1.insert(0,seq1[row_start-1])
      list2.insert(0,seq2[col_start-1])
      Max = table[row_start - 1][col_start - 1] # update the Max value
      row_start -=1
      col_start -=1
      

    elif case2 == table[row_start][col_start]: # top 2 bottom
      list1.insert(0,seq1[row_start-1])
      list2.insert(0,"-")
      Max = table[row_start - 1][col_start]
      row_start -=1
      
    
    elif case3 == table[row_start][col_start]: # left 2 right
      list1.insert(0,"-")
      list2.insert(0,seq2[col_start-1])
      Max= table[row_start][col_start - 1]
      col_start -=1
      

  for i in range(len(list2)):
    if list1[i] == list2[i]:
      blank.append("|")
    else:
      blank.append(' ')
  print("\n")
  print("The best alignmnet is: ")
  print(*list1)
  print(*blank)
  print(*list2)
  print("\n")
```

```python
def TraceBack(seq1, seq2, match=1, mismatch=-1, gap=-1):
  table = [[0]*(len(seq2)+1) for i in range(len(seq1)+1)] # create initial table with 0

# ## start update the first column and the first row

  for i in range(len(seq2)+1): # update first row
    table[0][i] = table[0][i] + mismatch *i
  for i in range(len(seq1)+1): # update first column
    table[i][0] = table[i][0] + mismatch *i
  
## build the alignment table

  for rowid in range (1,len(seq1)+1):
    for colid in range(1,len(seq2)+1):
      if seq1[rowid-1] == seq2[colid-1]:
        case1 = table[rowid-1][colid-1] + match
      else:
        case1 = table[rowid-1][colid-1] + mismatch

      case2 = table[rowid-1][colid]+gap

      case3 = table[rowid][colid-1]+gap

      table[rowid][colid] = max([case1,case2,case3])


## the best alignmnet
  
  col_start = len(seq2)
  row_start = len(seq1)
  list1=[]

  list2=[]
  while col_start > 0 or row_start > 0:
    if seq1[row_start-1] == seq2[col_start-1]:
      case1 = table[row_start - 1][col_start - 1] + match
    else:
      case1 = table[row_start - 1][col_start - 1] + mismatch

    case2 = table[row_start - 1][col_start] + gap

    case3 = table[row_start][col_start - 1] + gap


    if case1 == table[row_start][col_start]:
      list1.insert(0,seq1[row_start-1])
      list2.insert(0,seq2[col_start-1])
      row_start -=1
      col_start -=1

    elif case2 == table[row_start][col_start]: # top 2 bottom
      list1.insert(0,seq1[row_start-1])
      list2.insert(0,"-")
      row_start -=1
    
    elif case3 == table[row_start][col_start]: # left 2 right
      list1.insert(0,"-")
      list2.insert(0,seq2[col_start-1])
      col_start -=1

  return [list1, list2]



## best alignment socre


    
```

```python
def ConstructMSA(Sequences,match=1, mismatch=-1, gap_latter=-2):
  import pandas as pd
  table = [[0]*(len(Sequences)) for i in range(len(Sequences))] # create initial table with 0
  for i in Sequences:
    a = sequences.index(i) # rowID
    for j in Sequences:
      b = sequences.index(j) # colID
      len_seq1 = len(i)
      len_seq2 = len(j)

      # make the seq equal length, not enough add space  
      if len_seq1 > len_seq2:
        spaceNum = len_seq1 - len_seq2
        j = j + spaceNum * " "

      elif len_seq2 > len_seq1:
        spaceNum = len_seq2 - len_seq1
        i = i + spaceNum * " "

      #calculate the score
      score = 0
      for k in range(len(i)):
        if i[k] == j[k]:
          score += match
        elif i[k] != j[k] and i[k] != " " and j[k] != " ":
          score += mismatch
        else:
          score += gap_latter
      table[a][b] = score

      if i == j: # same seq the value should not calculate. give 0
        table[a][b] = 0 
  pd_table = pd.DataFrame(table)
  print("Score matrix:")
  print(pd_table)
  print("\n")
  
  # find the best aggregate comparison score
  temp_sum = []
  for i in table:
    temp_sum.append(sum(i))
  
  template_seq = sequences[temp_sum.index(max(temp_sum))]

  # print(temp_sum.index(max(temp_sum))) # print out the largest row sum in index, for class example is index 1, means 'ATTTA' should be the template for global alignment
  # print(template_seq)

  seq_trim = sequences.copy()
  seq_trim.remove(template_seq) # romove the temlate
  global_alignment = [] # hold global alignment result from traceback
  for i in seq_trim:
    global_alignment.append(TraceBack(template_seq,i))

  pd_global_alignment = pd.DataFrame(global_alignment)
  print("Best global alignment result save in same row:")
  print(pd_global_alignment) # Best global alignment result save in same row
  print("\n")

  # prepare for multiseq alignment, need to found the longest temp seq from the every global alignment, set it as template

  template_length=[] # hold the template_seq length from the alignment result(with gap result)
  for i in range(len(global_alignment)):
    template_length.append(len(global_alignment[i][0]))

  Max_length_index = template_length.index(max(template_length)) # get the max template_seq length index(with gap result)

  Max_length_template = global_alignment[Max_length_index][0] # get the max length template_seq(with gap result)

  result = [] #to store the final result
  result.append(Max_length_template)

  # multiseq alignment

  temp = []
  pointer_for_alignemnt = 0
  for i in range(len(global_alignment)):
    for k in range(len(Max_length_template)): #run len(Max_length_template) times
      if pointer_for_alignemnt == len(global_alignment[i][0]):
        temp.append("-")
      elif Max_length_template[k] != global_alignment[i][0][pointer_for_alignemnt]:
        temp.append("-")
      elif Max_length_template[k] == global_alignment[i][0][pointer_for_alignemnt]:
        if pointer_for_alignemnt < len(global_alignment[i][0]):
          temp.append(global_alignment[i][1][pointer_for_alignemnt])
          pointer_for_alignemnt+=1

    result.append(temp)
    pointer_for_alignemnt = 0
    temp = []

  pd_result = pd.DataFrame(result)
  print("MSA Result: ")
  print(pd_result)
```

```python
sequences = ['CATTT','ATTTA','ATTG','GCATA']
ConstructMSA(sequences)

 
  
```
