**4.  In order to handle the case that no existing indices a & b can reduce the number of breakpoints, we need to reverse an increasing strip. We need implement the following function:**

---



ref: https://more-itertools.readthedocs.io/en/stable/api.html#more_itertools.consecutive_groups

```python
import more_itertools as mit


iterable = [0, 1, 2, 5, 6, 7, 3, 4, 8]
[list(group) for group in mit.consecutive_groups(iterable)]
```

```python
def find_increasing_strip(List): # identify the indices a and b that define the second of increasing strips
  import more_itertools as mit
  increasing_strip_temp=[list(group) for group in mit.consecutive_groups(List)]
  # this tool treat single num as a increasing_strip so we need to remove it from result
  increasing_strip = []
  for i in range(len(increasing_strip_temp)):
    if len(increasing_strip_temp[i])>1:
      increasing_strip.append(increasing_strip_temp[i])

  a = increasing_strip[1][0]
  b = increasing_strip[1][len(increasing_strip[1])-1]
  
  return [List.index(a),List.index(b)]
```

**Algorithm 1: SimpleReversalSort()**

---



```python
def SimpleReversalSort(List):
  reversals = 0
  for i in range(len(List)):
    position = i
    if List.index(i) != position:
      reversals +=1
      List = perform_reversal(List, i, List.index(i))
  return [List,reversals]
```

**Algorithm 3: ImprovedBreakPointReversalSort()**

```python
def find_decreasing_strip(List): # identify the indices a and b that define the second of increasing strips
  List.reverse()
  import more_itertools as mit
  increasing_strip_temp=[list(group) for group in mit.consecutive_groups(List)]
  # this tool treat single num as a increasing_strip so we need to remove it from result
  decreasing_strip = []
  for i in range(len(increasing_strip_temp)):
    if len(increasing_strip_temp[i])>1:
      decreasing_strip.append(increasing_strip_temp[i])

  if decreasing_strip != []:
    a = decreasing_strip[0][0]
    b = decreasing_strip[0][len(decreasing_strip[0])-1]
    List.reverse()
    return [List.index(b),List.index(a)]
  else: 
    List.reverse()
    return [-1,-1]
```

```python
def ImprovedBreakPointReversalSort(List):
  reversals = 0
  BreakPoint = count_breakpoints(List)

  while BreakPoint != 0:
    if find_decreasing_strip(List) != [-1,-1]:
      temp_reversal  = find_best_reversal(List)

    else:
      temp_reversal = find_increasing_strip(List)
    
    List = perform_reversal(List,temp_reversal[0],temp_reversal[1])
    reversals+=1
    BreakPoint = count_breakpoints(List)

  return [List, reversals]
```
