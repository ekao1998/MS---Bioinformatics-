**Question 5. Reading spike protein sequences**

```python
from Bio import SeqIO
spike = []
spike_id = []
for seq_record in SeqIO.parse("lab01_spike_proteins.fasta", "fasta"):
  spike.append(seq_record.seq)
  spike_id.append(seq_record.id)

# print(spike)
# print(spike_id)
# print(spike[0][450:500])

spike_dict = {}

for i in range(len(spike)):
  spike_dict[spike_id[i]] = str(spike[i][450:500]) # I use biopython to extract the seq, so the seq is in Seq format, I use str to change back to string format

print(spike_dict)

```
