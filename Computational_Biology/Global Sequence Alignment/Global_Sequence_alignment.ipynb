```python
def alignment_table_updated(seq1,seq2):
  import pandas as pd
  print("Seq 1: ",seq1," with lenth ",len(seq1))
  print("Seq 2: ",seq2," with lenth ",len(seq2))
  print("The initial table with dimention ",len(seq1)+1, " and ", len(seq2)+1," for the sequence alignment is: ")
  table = [[0]*(len(seq2)+1) for i in range(len(seq1)+1)]
  for i in range(len(seq2)+1): # update first row
    table[0][i] = table[0][i] -i
  for i in range(len(seq1)+1): # update first column
    table[i][0] = table[i][0] -i
  pd_table = pd.DataFrame(table)
  print(pd_table)

alignment_table_updated("AATTATATT","ACGTTAT")
```

**Part2**

```python
def buildNWTable(seq1, seq2, match=1, mismatch=-1, gap=-1):
  import pandas as pd
  print('Seq1: ',seq1)
  print('Seq2: ',seq2)
  table = [[0]*(len(seq2)+1) for i in range(len(seq1)+1)] # create initial table with 0
  print("The initial alignment table is: ")
  pd_table = pd.DataFrame(table)
  print(pd_table)
  print("\n")
  for i in range(len(seq2)+1): # update first row
    table[0][i] = table[0][i] + mismatch *i
  for i in range(len(seq1)+1): # update first column
    table[i][0] = table[i][0] + mismatch *i

  for rowid in range (1,len(seq1)+1):
    for colid in range(1,len(seq2)+1):
      if seq1[rowid-1] == seq2[colid-1]:
        case1 = table[rowid-1][colid-1] + match
      else:
        case1 = table[rowid-1][colid-1] + mismatch

      case2 = table[rowid-1][colid]+gap

      case3 = table[rowid][colid-1]+gap

      table[rowid][colid] = max([case1,case2,case3])
  pd_table = pd.DataFrame(table)
  print("The final alignment table is:")
  print(pd_table)


buildNWTable("TTGCT","CTTCCT")
```
