**Problem 1: Compute the Number of Times a Pattern Appears in a Text**

```python
import time
def  PatternCount(text, pattern):
    count = 0
    for i in range((len(text) - len(pattern)+1)): # index start from 0, and desides how many runs
        if text[i:(len(pattern)+i)] == pattern: # try to find the pattern start from index 0
            count +=1
    # print(count)
    return count # if I use print not return, it will be class "noneType", I cant use noneType in other function


# print("Enter text: ")
# Text = input()
# print("Enter patterns: ")
# Pattern = input()

print(PatternCount("GCGCG","GCG"))
```

**Problem 2: Find the Most Frequent Words in a String**

```python
def FrequentWords(text, k_mer):
  output = [] # to store all the possible in ans list but there must be repeat seq 
  max = 0 # initial max
  for i in range((len(text) - k_mer +1)):
    seq = text[i:(k_mer+i)] # this for loop will get all possible seq(k-mer)
    compare = PatternCount(text, seq) # int
    if compare > max: # means need to remove all the previous stored seq
      output.clear()
      max = PatternCount(text, seq)
      output.append(seq)
    elif compare == max:
      output.append(seq)

    final = set(output) # remove repeat seq
  return final

print(*FrequentWords("ACGTTGCATGTCGCATGATGCATGAGAGCT", 4)) # now data stored in set want to access need to change back to list
#can put a * sign to remove the bracket or wirte a for loop to interate the ans
# a = list(FrequentWords("ACGTTGCATGTCGCATGATGCATGAGAGCT", 4))
# print(a[0])
```

**Problem 3: Find the Reverse Complement of a String**

```python

def rComplement(text):
  output = []
  DNA_dict = {
      "A":"T",
      "C":"G",
      "T":"A",
      "G":"C"
  }
  for i in text:
    output.append(DNA_dict[i])
  
  output.reverse()
  return "".join(output)

print(rComplement("AAAACCCGGT"))
      
```
