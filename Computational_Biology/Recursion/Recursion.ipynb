**Q1**

```python
def factorial(n):
  if n <0: 
    return "incorrect input"
  elif n==0:
    return 1

  else:
    return factorial(n-1)*n

print(factorial(0)) #1
print(factorial(1)) #1
print(factorial(2)) #2
print(factorial(3)) #6
print(factorial(6)) #720
print(factorial(-1)) #incorrect input
```

*   set n = 2
*   first run --> factorial(2) no "if" qualify, go into else --> factorial(2-1) * 2
* 2nd run --> factorial(1) no "if" qualify, go into else --> factorial(1-1) * 1
* 3rd run --> factorial(0) qualified, return 1
* Ans: 1 * 1 * 2 --> 2

```python
def get_sum(n):
  if n <0: 
    return "incorrect input"
  elif n ==0: 
    return 0
  else:
    return get_sum(n-1) + n

print(get_sum(2))
print(get_sum(10))
print(get_sum(0))
```

```python
def get_maximum(n):
    if len(n) == 1:
        return n[0]
    else:
        m = get_maximum(n[1:])
        if m>n[0]:
          return m
        else: # this one is to update the return value if m<n[0], become the new max num
          return n[0]


print(get_maximum([2,6,1,3,4,7]))
print(get_maximum([100,6,1,3,4,3])) # for this one first m=3,[4,3] compare to 4,so m = 4, compareto [3,4,3]
#m = 4, compare to [1,3,4,3], m = 4, compare to [6,1,3,4,3], m = 6, compare to [100,6,1,3,4,3], m = 100


```

* the maximum is either the first value in the list or the maximum of the rest of the list --> so the base case is list[0], if list[0] not the max then move the list by 1 -->  list[1:]. First break the list into the list only have 1 value:
* 2,6,1,3,4,7
* 6,1,3,4,7
* 1,3,4,7
* 3,4,7
* 4,7
* 7

* so now the return value m ==7  and recurse back compare to the n[0], the if else statement at the bottom
* 4,7 --> n[0] = 4, m = 7, m bigger no update 
* 3,4,7 --> n[0] = 3, m = 7, m bigger no update 
* .
* .
* 2,6,1,3,4,7 --> n[0]=2, m = 7, m bigger no update

```python
def FrequentWord_first_version(text, pattern, index): # set initial index to 0
  if len(text) - index  < len(pattern): # text can't be shorter than pattern
    return "end"
  elif text[index : len(pattern)+index] == pattern: #sliding window concept
    print(index) # if found print the index 
    return FrequentWord_first_version(text, pattern, index + 1) # and go in to next round to check all possible sliding window
  else:
    return FrequentWord_first_version(text, pattern, index + 1)

print(FrequentWord_first_version('GATATATGCATATACTT', 'ATAT', 0))
```

```python
def FrequentWord(text, pattern, index, location = []): # set initial index to 0, found can define an empty list like this, recursion won't reset the list
  if len(text) - index  < len(pattern): # text can't be shorter than pattern
    return location
  elif text[index : len(pattern)+index] == pattern: #sliding window concept
    location.append(index) # if found print the index 
    return FrequentWord(text, pattern, index + 1) # and go in to next round to check all possible sliding window
  else:
    return FrequentWord(text, pattern, index + 1)

print(FrequentWord('GATATATGCATATACTT', 'ATAT', 0))
```
