```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import numpy as np
```

```python
X = np.array([[0.5],[0.8],[0.2],[3],[6],[7],[4]])
y = np.array([0,0,0,1,1,1,1])

LDA_clf = LinearDiscriminantAnalysis(store_covariance=True)
LDA_clf.fit(X,y)
```

```python
print("Q1. Prior probability of classes: ",LDA_clf.priors_)
```

```python
print("Q3a. Mean of classes: ",LDA_clf.means_)
```

```python
print("Q3b. Variance: ",LDA_clf.covariance_)
```
