```python
#define gaussian prob distribution
import numpy as np
def gaussion_prob(x, u, sigma):
 prob = 1/(sigma*np.sqrt(2*np.pi))*np.exp(-(x-u)**2/(2*sigma**2))
 return prob

#set feature value
feature = 0.8

#get posterior probability of class 0 (assume the variances of classes are same)
pooled_variance = 1.45
priori_class0 = 3/7
posterior_class0 = gaussion_prob(x=feature, u=0.5, sigma = np.sqrt(pooled_variance)) * priori_class0
print("posterior_class0: ",posterior_class0)

#get posterior probability of class 1 (assume the variances of classes are same)
priori_class1 = 4/7
posterior_class1 = gaussion_prob(x=feature, u=5, sigma = np.sqrt(pooled_variance)) * priori_class1
print("posterior_class1: ",posterior_class1)
#determine the class based on the best posterior probability
```
